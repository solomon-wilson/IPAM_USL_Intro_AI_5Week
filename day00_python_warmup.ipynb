## 5 · Conditionals — the dispatch decision

`if` / `elif` / `else` runs exactly one branch. Comparisons like `>` and `==` answer True or False. SLPTA's rule of thumb: **over 15 minutes → pre-warn passengers.**

**Predict before running:** `delay_minutes` is 25 — which of the three messages prints?

```python
if delay_minutes > 15:
    print("PRE-WARN passengers - significant delay")
elif delay_minutes > 5:
    print("MONITOR - minor delay")
else:
    print("ON TIME - no action needed")
```

```python
# --- Exercise 5 (change one value) ------------------------------------------
threshold = 15   # TODO: change me to 30 - does the 25-minute delay still trigger a pre-warn?

if delay_minutes > threshold:
    print(f"PRE-WARN: {delay_minutes} min exceeds the {threshold}-min threshold")
else:
    print(f"No pre-warn: {delay_minutes} min is within the {threshold}-min threshold")
```

## 6 · Loops — the same work for every trip

`for` repeats a block once per item. Combined with `if`, it answers questions like *how many trips this week needed a pre-warn?*

**Predict before running:** the delays are `[25, 5, 0, 41, 12, 8]` — how many are above 15?

```python
late_count = 0
for delay in delays:
    if delay > 15:
        late_count = late_count + 1

print("Trips needing a pre-warn this week:", late_count)
```

```python
# --- Exercise 6 (fill one blank) -------------------------------------------
total_minutes = 0
for delay in delays:
    total_minutes = total_minutes + 0   # TODO: change the 0 so each trip's delay is added

print("Total minutes lost this week:", total_minutes)
```

## 7 · Functions — write the rule once, use it everywhere

`def` packages logic behind a name; `return` hands back the answer. A function is a standard operating procedure: written once, applied to any trip.

**Predict before running:** what will `delay_status(12)` return — and `delay_status(3)`?

```python
def delay_status(delay_minutes):
    """Return the dispatch action for a given delay in minutes."""
    if delay_minutes > 15:
        return "PRE-WARN"
    elif delay_minutes > 5:
        return "MONITOR"
    else:
        return "ON TIME"

print("R12 at 25 min ->", delay_status(25))
print("R7  at 12 min ->", delay_status(12))
print("R3  at  3 min ->", delay_status(3))
```

## 8 · Putting it together — your first dispatch rule

All eight blocks at once: a list of trip records, one rule function, one loop. **This is a rule-based classifier — and you wrote the rule.**

**Predict before running:** which of the four routes get PRE-WARN?

```python
trips = [
    {"route_id": "R12", "delay_minutes": 25},
    {"route_id": "R7",  "delay_minutes": 3},
    {"route_id": "R3",  "delay_minutes": 18},
    {"route_id": "R21", "delay_minutes": 0},
]

for trip in trips:
    action = delay_status(trip["delay_minutes"])
    print(f"{trip['route_id']:>3}: {trip['delay_minutes']:>2} min -> {action}")
```
